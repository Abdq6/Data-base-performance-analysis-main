```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df5 = pd.read_csv('./confidence_5/delete_latencies_1000_records_1_bytes.csv')
df10 = pd.read_csv('./confidence_10/delete_latencies_1000_records_1_bytes.csv')
df15 = pd.read_csv('./confidence_15/delete_latencies_150_records_1_bytes.csv')
df20 = pd.read_csv('./confidence_20/delete_latencies_1000_records_1_bytes.csv')
```

```python
df5.head()
```

```python
df5 = df5.melt()
df10 = df10.melt()
df15 = df15.melt()
df20 = df20.melt()

frames = [df5, df10, df15, df20]
colors = ['green', 'blue', 'black', 'red']
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
plt.style.use('bmh')
for frame, color, i, name in zip(frames, colors,[1,2,3,4], [5,10,15,20]):
    sns.lineplot(data=frame[:500*i], x='variable', y='value', marker='o',color=color, label=name)

ax.set_xticklabels(np.arange(1,101,2))
plt.xticks(np.arange(0,101,2))
plt.xticks(rotation=45,fontsize=10,style='italic')
plt.xticks(rotation=45,fontsize=10,style='italic')
plt.ylim([0,0.6])
plt.xlim([-1,102])
plt.legend(fancybox=True, shadow=True, facecolor='white', edgecolor='black',title='IloÅ›Ä‡ powtÃ³rzeÅ„',title_fontproperties={'weight':'bold'})
plt.ylabel('OpÃ³Åºnienie [ms]')
plt.xlabel('Iteracja eksperymentu')
plt.savefig('./conf.svg',bbox_inches="tight")
plt.show()
```
